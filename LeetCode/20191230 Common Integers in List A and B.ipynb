A and B are two sorted lists containing integers, we want to get the common element of A and B. Note that we can only access one element each time from A or B.

```python
A = [1,3,3,5,5,7,9]
B = [1,2,3,3,4,5,5,6,8]
```

```python
def CommonInt(A,B):
    # Time Complexity: O(n)
    pa,pb,res = 0,0,[]
    while pa<len(A) and pb<len(B):
        if A[pa]==B[pb]:
            temp = A[pa]
            res.append(temp)
            pa+=1
            pb+=1
            while A[pa]==temp:
                pa+=1
            while B[pb]==temp:
                pb+=1
        elif A[pa]>B[pb]:
            pb+=1
        else:
            pa+=1
    return res
```

```python
CommonInt(A,B)
```
