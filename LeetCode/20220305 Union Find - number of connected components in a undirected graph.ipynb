```python
def countComponents(n, edges):
    parent = [i for i in range(n)]
    rank = [i for i in range(n)]
    
    #given n1, find the root parent
    def find(n1):
        res = n1
        
        while res != parent[res]: #keep updating while it's not a root parent
            parent[res] = parent[parent[res]] #update towards the root parent
            res = parent[res] #keep moving towards the root parent
        
        return res
    
    #union node n1 and n2 into 1 undirected graph, return 1 if union successfully else 0
    #used to generate undirected graph from scratch
    #used to merge several undirected graphs
    def union(n1, n2):
        #parent 1 & 2
        p1, p2 = find(n1), find(n2)
        
        #if n1 and n2 belong to the same parent, don't union and return 0
        if p1 == p2:
            return 0
        
        #by defalut, the parent node with larger rank should be root
        if rank[p1]<rank[p2]:
            parent[p1] = p2
            rank[p2] += rank[p1]
        else:
            parent[p2] = p1
            rank[p1] += rank[p2]
        #return 0 if union is successful
        return 0
    
    res = n
    for n1,n2 in edges:
        res -= union(n1,n2)
    
    return res
```

```python
n = 5
edges = [ [0,1],[1,2],[3,4] ]
countComponents(n, edges)
```
