```python
def findTargetSumWays(nums, S):
    """
    :type nums: List[int]
    :type S: int
    :rtype: int
    """
    n = len(nums)
    if nums==None or n==0:
        if S==0:
            return 0
    if sum(nums)<S:
        return 0
    
    nums.sort()
    nzeros = nums.count(0)
    nums=nums[nzeros:]
    target = (sum(nums)-S)/2.0
    if target%1!=0:
        return 0
    elif target==0:
        return pow(2,nzeros)
    else:
        target = int(target)
        DP = [[0 for i in range(target+1)] for j in range(len(nums)+1)]
        for i in range(1,len(nums)+1):
            cur = nums[i-1]
            for j in range(1,target+1):
                if j<cur:
                    DP[i][j]=DP[i-1][j]
                elif j==cur:
                    DP[i][j]=DP[i-1][j]+1
                else:
                    DP[i][j]=DP[i-1][j]+DP[i-1][j-cur]
        return DP[-1][-1]*pow(2,nzeros)
```

```python
findTargetSumWays([0,1,1,1,1,1],1)
```
