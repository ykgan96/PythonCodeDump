```python
import numpy as np
```

```python
def factorial(n):
    if n == 0:
        return 1
    prod = 1
    for i in range(1,n+1):
        prod *=i
    return prod

def C(n,k):
    return factorial(n)/(factorial(k)*factorial(n-k))

def P(n,k):
    return factorial(n)/factorial(n-k)
```
