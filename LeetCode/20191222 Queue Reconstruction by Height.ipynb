```python
def reconstructQueue(people):
    """
    :type people: List[List[int]]
    :rtype: List[List[int]]
    """
    if len(people)<=1:
        return people
        
    res = []
    height = [people[0][0]]
    dic = {people[0][0]:[(0,people[0][1])]}
        
    def find_ind(l,k):
        if l[0][1]>k:
            return 0
        elif l[-1][1]<k:
            return len(l)
        for i in range(1,len(l)):
            if l[i-1][1]<k and l[i][1]>k:
                return i
                         
    for i in range(1,len(people)):
        cur = people[i][0]
        if cur in height:
            index = find_ind(dic[cur],people[i][1])
            dic[cur].insert(index,(i,people[i][1]))
        else:
            dic[cur] = [(i,people[i][1])]
            height.append(cur)
        
    for h in sorted(height)[::-1]:
        for ind in dic[h]:
            res.insert(people[ind[0]][1],people[ind[0]])
                
    return res
```
