```python
def ZerosInFactorial(n):
    if n<5: return 0
    return n//5 + ZerosInFactorial(n//5)
```

```python
ZerosInFactorial(100)
```

# Random Number Generator

```python
import numpy as np
def RNG3070():
    # base random number generator
    # generate A with probablity 30% and B with 70% probability
    return 'A' if np.random.random() < 0.3 else 'B'
```

```python
def RNG5050():
    # advanced random number generator
    # generate A with probablity 50% and B with 50% probability
    cur_rng = RNG3070() + RNG3070()
    while cur_rng != 'AB' and cur_rng != 'BA':
        cur_rng = RNG3070() + RNG3070()
    return 'A' if cur_rng == 'AB' else 'B'
```

```python
def RNG5050_multiple(n):
    # n: number of randoms to be generated
    return [RNG5050() for _ in range(n)]
```

```python
RNG5050_multiple(10)
```
