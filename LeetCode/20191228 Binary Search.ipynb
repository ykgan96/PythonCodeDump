```python
def naive(nums,num):
    if num<nums[0]:
        nums[0]=num
    if num>nums[-1]:
        nums.append(num)
    for i in range(1,len(nums)):
        if nums[i-1]<num and num<=nums[i]:
            nums[i]=num
    return nums
```

```python
naive([1,3,5,7,9],6)
```

```python
naive([3,5,7,9,11,13],10)
```

```python
naive([3,5,7,9,11,13],15)
```

```python
def BS(nums,num):
    i,j = 0,len(nums)
    while i!=j:
        m = (i+j)//2
        if nums[m]<num:
            i = m+1
        else:
            j = m
    if i==len(nums):
        nums.append(num)
    else:
        nums[i]=num
    return nums
```

```python
BS([1,3,5,7,9],1)
```

```python
BS([1,3,5,7,9],6)
```

```python
BS([1,3,5,7,9],10)
```

```python
BS([3,5,7,9,11,13],1)
```

```python
BS([3,5,7,9,11,13],8)
```

```python
BS([3,5,7,9,11,13],14)
```
