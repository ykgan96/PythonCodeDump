```python
def SubsetSum(total,NumList):
    n = len(NumList)
    DP = [[True for i in range(total+1)] for j in range(n)]
    
    for j in range(1,total+1):
        if j!=NumList[0]:
            DP[0][j]=False
    
    for i in range(1,n):
        for j in range(1,total+1):
            cur = NumList[i]
            if j<cur:
                DP[i][j]=DP[i-1][j]
            else:
                DP[i][j] = (DP[i-1][j]==True or DP[i-1][j-cur]==True)
    #return DP[n-1][total]
    return [i for i in range(1,total+1) if DP[n-1][i]==True]
```

```python
SubsetSum(sum([2,3,7,8,10]),[2,3,7,8,10])
```
