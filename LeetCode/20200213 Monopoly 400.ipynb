# 1Dice

```python
arr = [0]*401
arr[0] = 1
arr[1] = 1/6*arr[0]
arr[2] = 1/6*arr[0]+1/6*arr[1]
arr[3] = 1/6*arr[0]+1/6*arr[1]+1/6*arr[2]
arr[4] = 1/6*arr[0]+1/6*arr[1]+1/6*arr[2]+1/6*arr[3]
arr[5] = 1/6*arr[0]+1/6*arr[1]+1/6*arr[2]+1/6*arr[3]+1/6*arr[4]
arr[6] = 1/6*arr[0]+1/6*arr[1]+1/6*arr[2]+1/6*arr[3]+1/6*arr[4]+1/6*arr[5]
```

```python
for i in range(7,401):
    arr[i] = 1/6*arr[i-6]+1/6*arr[i-5]+1/6*arr[i-4]+1/6*arr[i-3]+1/6*arr[i-2]+1/6*arr[i-1]
```

# 2 Dices

```python
p = [0,0,1/36,2/36,3/36,4/36,5/36,6/36,5/36,4/36,3/36,2/36,1/36]
```

```python
prob = [0]*401
prob[0] = 1
for i in range(2,401):
    for j in range(2,13):
        if i-j>=0:
            prob[i] += prob[i-j]*p[j]
prob
```

# Time Summary

```python
import pandas as pd
df = pd.read_csv('time.csv',index_col=0)
df = df.dropna()
df
```

```python
def get_hour(row):
    if row['timestamp'][1]==':': 
        return row['timestamp'][0]
    else:
        return row['timestamp'][:2]
```

```python
df['hour'] = df.apply(lambda row: get_hour(row),axis=1)
df
```

```python
df.groupby('hour').count()
```

```python
count_list = df.groupby('hour').count().timestamp.tolist()
max(count_list)-min(count_list)
```
