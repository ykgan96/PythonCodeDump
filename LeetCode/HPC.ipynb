```python
import concurrent.futures as cf
import requests
import time
import json

def download_from_url(url, fname):
    try:
        t0 = time.time()
        res = requests.get(url)  # get content
        # add a function to save file here
        with open(fname, "w") as f:
            json.dump(res.text, f)
    except Exception as e:
        print(f"Error when downloading from {url}: {e}")
    finally:
        t1 = time.time()
        print(f"Duration: {t1-t0:.4f}s, URL: {url}")
        
urls = [
    "https://www.example.com",
    "https://www.python.org",
    "https://www.github.com",
    'https://api.github.com'
]

start_time = time.time()
with cf.ThreadPoolExecutor() as executor:
    futures = [executor.submit(download_from_url, urls[i], f"{i}.json") for i in range(len(urls))]
    
end_time = time.time()
print(f"Downloaded all data in {end_time - start_time:.4f} seconds")
```

```python
import concurrent.futures
import time
import requests

# Function to download a webpage
def download_page(url):
    s = time.time()
    response = requests.get(url)
    return f"{url} - {round(time.time()-s,4)}s - {len(response.content)} bytes"

# List of URLs to download
urls = [
    "https://www.example.com",
    "https://www.python.org",
    "https://www.github.com",
]

# Using ThreadPoolExecutor to download pages concurrently
start_time = time.time()

with concurrent.futures.ThreadPoolExecutor() as executor:
    futures = [executor.submit(download_page, url) for url in urls]
    for future in concurrent.futures.as_completed(futures):
        print(future.result())

end_time = time.time()
print(f"Downloaded all pages in {end_time - start_time:.2f} seconds")
```
