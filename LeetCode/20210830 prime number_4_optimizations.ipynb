```python
def is_prime(k):
    count = 0
    for i in range(1,int(k**0.5)+1):
        if k%i==0:
            count += 1
            if k/i!=i:
                count += 1
    return count==2
    
def prime_numbers(n):
    # TC: O(n^1.5)
    if n==0: return []
    
    res = []
    for k in range(1,n+1):
        if is_prime(k):
            res += [k]
    return res
```

```python
def prime_numbers_improved_1(n):
    # TC before n**0.5 : n/2+n/3+n/5+n/7+n/11+...+n/(the last prime number before n)
    # TC: O(nlog(logn))
    if n==0: return []
    if n==1: return []
    if n==2: return [2]
    
    status = [True for _ in range(n+1)]
    status[0], status[1] = False, False
    res = []
    for i in range(2, n+1):
        if status[i]==True:
            res += [i]
            temp = i*2
            while temp<=n:
                status[temp] = False
                temp += i   
    return res
```

```python
prime_numbers_improved_1(50)
```

```python
def prime_numbers_improved_2(n):
    # TC before n**0.5 : n/2+n/3+n/5+n/7+n/11+...+n/(the last prime number before n**0.5)
    # TC after n**0.5: 1+1+1+...+1
    # TC: O(nlog(log(n**0.5))) + O(n-n**0.5) = O(nlog(log(n**0.5)))
    if n==0: return []
    if n==1: return []
    if n==2: return [2]
    
    status = [True for _ in range(n+1)]
    status[0], status[1] = False, False
    res = []
    for i in range(2, int(n**0.5)+1):
        if status[i]==True:
            res += [i]
            temp = i*2
            while temp<=n:
                status[temp] = False
                temp += i
    
    #optimize
    for i in range(int(n**0.5)+1, n+1):
        if status[i]==True:
            res += [i]
    
    return res
```
