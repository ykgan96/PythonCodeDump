# Max Sub Square Matrix + Count Square Submatrices with All Ones

```python
def SSM(matrix):
    # matrix: List[List[int]]
    m, n = len(matrix), len(matrix[0])
    mat = [[0 for i in range(n+1)] for j in range(m+1)]
    for i in range(m):
        for j in range(n):
            if matrix[i][j]==0:
                mat[i+1][j+1]=0
            else:
                mat[i+1][j+1]=min(min(mat[i][j+1],mat[i+1][j]),mat[i][j])+1
    return max([max(item) for item in mat]), sum([sum(item) for item in mat])
```

```python
matrix =[[0,1,1,1],
         [1,1,1,1],
         [0,1,1,1]]
```

```python
MSSM, CSSAO = SSM(matrix)
print('Max Sub Square Matrix: '+str(MSSM)+'*'+str(MSSM))
print('Count Square Submatrices with All Ones: '+str(CSSAO))
```

# Method 2

```python
def Mat(matrix):
    m,n = len(matrix),len(matrix[0])
    for i in range(1,m):
        for j in range(1,n):
            if matrix[i][j]==0:
                continue
            else:
                matrix[i][j]=min(min(matrix[i-1][j],matrix[i][j-1]),matrix[i-1][j-1])+1
    return max([max(item) for item in matrix]), sum([sum(item) for item in matrix])
```

```python
MSSM, CSSAO = Mat(matrix)
print('Max Sub Square Matrix: '+str(MSSM)+'*'+str(MSSM))
print('Count Square Submatrices with All Ones: '+str(CSSAO))
```
