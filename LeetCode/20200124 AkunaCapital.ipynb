# Binary Representation

```python
def binary(num):
        temp,arr,count = num,[],0
        while num!=0:
            if num%2==1: count+=1
            arr.append(str(num%2))
            num = num//2
        BinaryRep = ''.join(arr)[::-1]
        return [count,int(BinaryRep),temp]
```

```python
binary(8)
```

```python
def rearrange(elements):
    # Write your code here
    if not elements or len(elements)==1: return elements
    
    res = []
    for e in elements:
        res.append(binary(e))
    res.sort()
    
    for r in res:
        print(r)
```

```python
rearrange([5,10,3,7,14])
```

# Change Sort

```python
def changedSort(s):
    if not s or len(set(s))==1: return s
    ss, total_len, remain, s_sort = '', len(s), [], sorted(list(s))
    
    while len(ss)<total_len:
        for i in range(0,len(s_sort)):
            if i==0:
                ss+=s_sort[i]
                continue
            if s_sort[i]!=s_sort[i-1]:
                ss+=s_sort[i]
            else:
                remain.append(s_sort[i])
        s_sort = remain
        remain = []
        if len(set(s_sort))==1:
            ss+=''.join(s_sort)
            break
            
        if len(s_sort)>0:
            ss+=s_sort[-1]
            for i in range(-2,-len(s_sort)-1,-1):
                if s_sort[i]!=s_sort[i+1]:
                    ss+=s_sort[i]
                else:
                    remain.append(s_sort[i])
            remain.sort()
            s_sort = remain
            remain = []
            if len(set(s_sort))==1: 
                ss+=''.join(s_sort)
                
    return ss
```

```python
changedSort('abababababbabcef')
```
