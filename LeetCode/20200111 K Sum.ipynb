```python
def KSum(nums,target,K,result,resList):
    if len(nums)<K or K<2: return
    
    if K==2:
        l,r = 0,len(nums)-1
        while l<r:
            if nums[l]+nums[r]==target:
                resList.append(result+[nums[l],nums[r]])
                while l<r and nums[l]==nums[l+1]: l+=1
                while l<r and nums[r]==nums[r-1]: r-=1
                l+=1
                r-=1
            elif nums[l]+nums[r]>target:
                r-=1
            else:
                l+=1
        return
    else:
        for i in range(len(nums)-K+1):
            if nums[i]*K>target or nums[-1]*K<target: break
            if i==0 or (i>0 and nums[i]!=nums[i-1]):
                KSum(nums[i+1:],target-nums[i],K-1,result+[nums[i]],resList)
    return
```

# FourSum

```python
def FourSum(nums,target):
    nums.sort()
    resList = []
    KSum(nums,target,4,[],resList)
    return resList
```

```python
FourSum([1, 0, -1, 0, -2, 2],0)
```

# FiveSum

```python
def FiveSum(nums,target):
    nums.sort()
    resList = []
    KSum(nums,target,5,[],resList)
    return resList
```

```python
FiveSum([1, 2, 3, 5, -4, 6, 8, -10, -12, 15, 0, -1, 0, -2, 2],0)
```
