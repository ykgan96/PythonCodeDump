```python
grid = ['eabcd', 'fghij', 'olkmn', 'trpqs', 'xywuv']
```

```python
def gridChallenge(grid):
    for i in range(len(grid)):
        grid[i] = sorted(list(grid[i]))
        
    grid = [*zip(*grid)]
    for row in grid:
        if list(row)!=list(sorted(row)):
            return 'NO'
    return 'YES'
```

```python
gridChallenge(grid)
```
