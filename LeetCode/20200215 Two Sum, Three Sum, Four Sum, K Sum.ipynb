# Two Sum

```python
def TwoSum(nums,target):
    if len(nums)<2: return []
    nums.sort()
    res = []
    l,r = 0,len(nums)-1
    while l<r:
        if nums[l]+nums[r]==target:
            res.append([nums[l],nums[r]])
            while l<r and nums[l]==nums[l+1]: l+=1
            while l<r and nums[r]==nums[r-1]: r-=1
            l+=1
            r-=1
        elif nums[l]+nums[r]<target:
            l+=1
        else:
            r-=1
    return res
```

```python
TwoSum([-1, 0, 1, 2, -1, -4],0)
```

# Three Sum

```python
def ThreeSum(nums,target):
    if len(nums)<3: return []
    nums.sort()
    res=[]
    for i in range(len(nums)):
        if i>0 and nums[i-1]==nums[i]:
            continue
        l,r = i+1,len(nums)-1
        while l<r:
            if nums[i]+nums[l]+nums[r]==target:
                res.append([nums[i],nums[l],nums[r]])
                while l<r and nums[l]==nums[l+1]: l+=1
                while l<r and nums[r]==nums[r-1]: r-=1
                l+=1
                r-=1
            elif nums[i]+nums[l]+nums[r]>target:
                r-=1
            else:
                l+=1
    return res
```

```python
ThreeSum([-1, 0, 1, 2, -1, -4],0)
```

# K Sum

```python
def kSum(nums,target,k,res,resList):
    # nums: an array containing sorted numbers
    if len(nums)<k or k<2: return []
    
    if k==2:
        l,r = 0, len(nums)-1
        while l<r:
            if nums[l]+nums[r]==target:
                resList.append(res+[nums[l],nums[r]])
                while l<r and nums[l]==nums[l+1]: l+=1
                while l<r and nums[r]==nums[r-1]: r-=1
                l+=1
                r-=1
            elif nums[l]+nums[r]<target:
                l+=1
            else:
                r-=1
        return
    else:
        for i in range(len(nums)):
            if nums[i]*k>target or nums[-1]*k<target: break
            if i>0 and nums[i]==nums[i-1]: continue
            kSum(nums[i+1:],target-nums[i],k-1,res+[nums[i]],resList)
    return 
```

# Four Sum

```python
def FourSum(nums,target):
    nums.sort()
    resList = []
    kSum(nums,target,4,[],resList)
    return resList
```

```python
FourSum([-1, 0, 1, 2, -1, -4],0)
```

```python
def FiveSum(nums,target):
    nums.sort()
    resList=[]
    kSum(nums,target,5,[],resList)
    return resList
```
