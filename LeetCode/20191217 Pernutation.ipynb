```python
def permutation(n):
    arr = [i for i in range(1,n+1)]
    res = []
    
    def DFS(arr,temp):
        if len(temp)==n:
            res.append(temp)
            
        for i in range(len(arr)):
            DFS(arr[:i]+arr[i+1:],temp+[arr[i]])
            
    DFS(arr,[])
    return res
```

```python
permutation(3)
```

```python
def permutation(arr):
    n=len(arr)
    res = []
    
    def DFS(arr,temp):
        if len(temp)==n and temp not in res:
            res.append(temp)
            
        for i in range(len(arr)):
            DFS(arr[:i]+arr[i+1:],temp+[arr[i]])
            
    DFS(arr,[])
    return res
```

```python
permutation(['l','d','d','l'])
```
