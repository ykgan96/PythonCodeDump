# DFS

```python
def WordSearch(word,board):
    """
    :type board: List[List[str]]
    :type word: str
    :rtype: bool
    """
    if not board or len(board)==0 or len(board[0])==0 or len(word)==0 or not word:
        return False

    m,n,l= len(board),len(board[0]),len(word)
    if m*n<l:
        return False
    
    def dfs(cur,i,j):
        if cur==word: return True
        if i not in range(m) or j not in range(n) or word[len(cur)]!=board[i][j]:
            return False
        cur += board[i][j]
        temp = board[i][j]
        board[i][j] = '#'
        res = dfs(cur,i+1,j) or dfs(cur,i-1,j) or dfs(cur,i,j+1) or dfs(cur,i,j-1)
        board[i][j] = temp
        return res
        
    for i in range(m):
        for j in range(n):
            if board[i][j]==word[0]:
                if dfs('',i,j):
                    return True
    return False
```

```python
board = [["b","a","a","b","a","b"],["a","b","a","a","a","a"],["a","b","a","a","a","b"],["a","b","a","b","b","a"],["a","a","b","b","a","b"],["a","a","b","b","b","a"],["a","a","b","a","a","b"]]
word = "aabbbbabbaababaaaabababbaaba"
```

```python
WordSearch(word,board)
```
