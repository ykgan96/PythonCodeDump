# Dynamic Programming with Fibonacci

```python
import numpy as np
```

Naive Algo

```python
def fib_naive(n):
    if n==1 or n==2:
        return 1
    else:
        return fib_naive(n-1)+fib_naive(n-2)
```

```python
fib_naive(30)
```

DP: memory list approach

```python
length = 30
memo = [np.nan]*(length+1)
def fib_memo(n,memo):
    if np.isnan(memo[n])==False:
        res = memo[n]
    if n==1 or n==2:
        memo[n] =1
        res = 1
    else:
        res = fib_memo(n-1,memo)+fib_memo(n-2,memo)
        memo[n]= res
    return res
fib_memo(length,memo)
```

DP: Bottom-up Approach

```python
def fib_bu(n):
    memo = [np.nan]*(n+1)
    memo[1], memo[2] = 1, 1
    for i in range(3,n+1):
        memo[i] = memo[i-1]+memo[i-2]
    return memo[n]
```

```python
fib_bu(30)
```
