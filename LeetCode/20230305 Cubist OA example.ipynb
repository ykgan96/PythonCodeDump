# Quater Transformation

```python
def get_quarter(date):
    mapping = [([1, 2, 3], 1),([4, 5, 6], 2),([7, 8, 9], 3),([10, 11, 12], 4)]
    
    for months, quarter in mapping:
        if date.month in months:
            return str(date.year)+"Q"+str(quarter)
```

```python
get_quarter(datetime.datetime.today())
```

# Returns sum of all odd factors of n

```python
def oddDivisorSum(n):
    #get rid of all even factors
    while n % 2 == 0:
        n //= 2
        
    #consider all the odd factors from 1 to sqrt(n)
    sum = 0
    
    for k in range(1, int(n ** .5) + 1, 2):
        if n % k == 0:
            sum += k
            print(k)
            if n // k != k: #consider its conterpart
                sum += n // k
                print(n//k)
                
    return sum
```
