```python
def BinaryRep(n):
    res = ''
    while n:
        res += str(n%2)
        n = n//2
    return res[::-1]
```

```python
BinaryRep(10)
```

```python
def BR(num):
    return [bin(num)[2:].count('1'),num]
```

```python
def WS(n,k):
    res = []
    for i in range(n):
        res.append(BR(i))
    return sorted(res)[k][1]
```

```python
WS(8,5)
```
