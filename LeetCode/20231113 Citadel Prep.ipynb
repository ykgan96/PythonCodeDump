# Q1: Missing Number

Given an array arr[] of size N-1 with integers in the range of [1, N], the task is to find the missing number from the first N integers.

Note: There are no duplicates in the list.

```python
def missing_num(arr):
    N = len(arr) + 1
    return N*(N+1)/2 - sum(arr)
```

```python
arr = [1, 2, 4, 6, 3, 7, 8]
missing_num(arr)
```

# Q2: Trapping Rain Water

Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

```python
def trap_rain_water(arr):
    n, n_water = len(arr), 0
    if n<=2: return 0
    
    # left[i]: the height of the highest bar to the left of index i
    left = [0]*n
    for i in range(1, n):
        left[i] = max(left[i-1], arr[i-1])
    #print(left)
    
    # right[i]: the height of the highest bar to the right of index i
    right = [0]*n
    for i in range(n-2, -1, -1):
        right[i] = max(right[i+1], arr[i+1])
    #print(right)
    
    for i in range(n):
        n_water += max(min(left[i],right[i])-arr[i], 0)
    
    return n_water
```

```python
arr = [0,1,0,3,0,2]
arr = [0,1,0,2,1,0,1,3,2,1,2,1]
arr = [4,2,0,3,2,5]
trap_rain_water(arr)
```

```python
def trap_rain_water_v2(arr):
    n = len(arr)
    if n<=2: return 0
    
    n_water, l, r, max_l, max_r = 0, 0, n-1, arr[0], arr[n-1]
    # n_water = max(0, min(max_l, max_r)-arr[i])
    while l<r:
        if max_l<max_r:
            n_water += (max_l - arr[l])
            l += 1
            max_l = max(max_l, arr[l])
        else:
            n_water += (max_r - arr[r])
            r -= 1
            max_r = max(max_r, arr[r])
    return n_water
```

```python
arr = [0,1,0,3,0,2]
arr = [0,1,0,2,1,0,1,3,2,1,2,1]
arr = [4,2,0,3,2,5]
trap_rain_water(arr)
```

# Q3: Find the Index of the First Occurrence in a String
Given two strings a and b, return the index of the first occurrence of a in b, or -1 if a is not part of b.

```python
def strStartIdx(a, b):
    for i in range(len(b)-len(a)+1):
        if b[i:i+len(a)]==a:
            return i
    return -1
```

```python
a, b = 'llo', 'hello'
strStartIdx(a, b)
```

# Q4: Hash Table with Hash Func

```python
class HashTable(object):
    def __init__(self, Max=100):
        self.Max = Max
        self.arr = [None for i in range(self.Max)]
        
    def hash_func(self, key):
        res = 0
        for s in key:
            res += ord(s)
        return res % self.Max
    
    def add(self, key, val):
        h = self.hash_func(key)
        self.arr[h] = val
        
    def get(self, key):
        h = self.hash_func(key)
        return self.arr[h]
    
    def __setitem__(self, key, val):
        self.add(key, val)
        
    def __getitem__(self, key):
        return self.get(key)
        
    def delete(self, key):
        h = self.hash_func(key)
        self.arr[h] = None
        
    def __delitem__(self, key):
        self.delete(key)
```

# Q6: Is a valid Binary Search Tree

```python
def validateBST(root, min_v, max_v):
    if not root: return True
    if root.val<=min_v or root.val>=max_v: return False
    return validateBST(root.left, min_v, root.val) and validateBST(root.right, root.val, max_v)
    
def isValidBST(root):
    return validateBST(root, -float('inf'), float('inf'))
```

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

left_child = TreeNode(val=1, left=None, right=None)
right_child = TreeNode(val=3, left=None, right=None)
root = TreeNode(2, left=left_child, right=right_child)
```

```python
isValidBST(root)
```
