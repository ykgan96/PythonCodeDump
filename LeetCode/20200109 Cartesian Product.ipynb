```python
dct = {'1':'abc','2':'def','3':'ghi','4':'jklm','5':'nopq','6':'rst','7':'uvw','8':'xyz'}
```

# DFS

```python
def dfs(s):
    n,res = len(s),[]
    
    def dfs_(cur):
        if len(cur)==n:
            res.append(cur)
            return
        temp = dct[s[len(cur)]]
        for x in temp:
            dfs_(cur+x)
    
    dfs_('')
    return res
```

```python
dfs('12')
```

```python
dfs('8283183635')
```

# BFS

```python
def ListProd(l1,l2):
    ans = []
    for x in l1:
        for y in l2:
            ans.append(x+y)
    return ans
```

```python
def bfs(s):
    if len(s)==0: return []
    if len(s)==1: return list(dct[s[0]])
    ans = ListProd(dct[s[0]],dct[s[1]])
    for i in range(2,len(s)):
        ans = ListProd(ans,dct[s[i]])
    return ans
```

```python
bfs('12')
```

```python
bfs('8283183635')
```
