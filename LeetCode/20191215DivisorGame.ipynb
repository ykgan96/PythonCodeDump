# Find Divisors

```python
def find_divisors(number):
    divisors = []
    for i in range(1,int(number**0.5)+1):
        if number%i==0:
            divisors+=[i,number//i]
    divisors = list(set(divisors))
    divisors.sort()
    return divisors
```

```python
find_divisors(100)
```

# Divisor Game

```python
def after_divisors(number):
    res = []
    for i in range(1,int(number**0.5)+1):
        if number%i==0:
            res+=[number-i,number-number//i]
    res = list(set(res))
    res.sort()
    return res[1:]
```

```python
after_divisors(4)
```

```python
def DivisorGame(N):
    if N==1 or N==3: return False
    if N==2: return True
    D = [-1 for i in range(N+1)]
    D[1], D[2], D[3] = False, True, False
    for i in range(4,N+1):
        d = after_divisors(i)
        for item in d:
            if D[item]==False:
                D[i]=True
        if D[i]==-1:
            D[i]=False
    return D
```
