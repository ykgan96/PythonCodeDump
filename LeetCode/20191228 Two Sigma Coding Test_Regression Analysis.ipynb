```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import statsmodels.api as sm
```

```python
df = pd.DataFrame([[70,95,34,46,10,50],[65,88,45,24,32,51],
                   [87,91,23,35,10,78],[67,101,34,55,15,88]]
                  ,columns = ['Town1','Town2','Town3','Town4','Town5','NYC'])
df
```

```python
columns = ['Town1','Town2','Town3','Town4','Town5','NYC']
','.join(columns)
```

```python
town_list = df.columns.tolist()
town_list.remove('NYC')
X = df[town_list]
y = df['NYC']
X =sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
model.tvalues.sort_values()[1:]
```

```python
import statsmodels.api as sm
q3, mse = 0.0, []
for town_i in df.columns.tolist()[:-1]:
    X = df[town_i] 
    y = df['NYC']
    X =sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    q3 += abs(model.params[1])
    pred = model.predict(X)
    cur_mse = np.sum((y-pred)**2)/len(pred)
    print(cur_mse,town_i)
    mse.append([cur_mse,town_i])
q3 = str(int(round(q3)))+','
q3
```
