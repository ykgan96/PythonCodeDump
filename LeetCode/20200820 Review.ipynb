```python
def BS(obj,nums):
    l, r = 0, len(nums)-1
    while l<r:
        m = (l+r)//2
        if nums[m]==obj:
            return m
        elif nums[m]>obj:
            r=m-1
        else:
            l=m+1
    if nums[l]==obj:
        return l
    else:
        return -1
```

```python
BS(2,[0,1,3])
```

```python
BS(3,[1,3,5,6,7,8,9])
```
