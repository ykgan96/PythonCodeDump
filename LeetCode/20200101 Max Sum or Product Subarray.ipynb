```python
def MaxSumSubarray(arr):
    # TC: o(n)
    if not arr: return False
    if len(arr)==1: return arr[0]
    Min = min(0,arr[0]) # 0 if not include the first element, arr[0] if include the first element
    Max = arr[0] #Max must include at least one element
    PreSum = arr[0]
    for i in range(1,len(arr)):
        PreSum += arr[i] #current sum
        Max = max(Max,PreSum-Min) # max A[j:i] = A[0:i] - min A[0:j] 
        Min = min(Min,PreSum) #update Min
    return Max
```

```python
MaxSumSubarray([-2,1,-3,4,-1,2,1,-5,4])
```

```python
def MaxProdSubarray(arr):
    res,imax,imin = arr[0],arr[0],arr[0]
    for i in range(1,len(arr)):
        if arr[i]<0:
            imax,imin = imin,imax
        imax = max(arr[i],imax*arr[i])
        imin = min(arr[i],imin*arr[i])
        res = max(res,imax)
    return res
```

```python
MaxProdSubarray([2,3,-2,4])
```

```python
def MaxProd(arr):
    DP = [[0,0] for _ in range(len(arr))]
    DP[0] = (arr[0],arr[0])
    res = arr[0]
    for i in range(1,len(arr)):
        prev = DP[i-1]
        imax = max(max(arr[i],arr[i]*prev[0]),arr[i]*prev[1])
        imin = min(min(arr[i],arr[i]*prev[0]),arr[i]*prev[1])
        DP[i] = [0,0]
        res = max(res,imax)
    return res
```

```python
MaxProd([2,3,-2,4])
```
