```python
import numpy as np
import pandas as pd
```

```python
# (L_i, W_i, H_i)
boxes = [
    [4,5,3],
    [2,3,2],
    [3,6,2],
    [1,5,4],
    [2,4,1],
    [1,2,2]
]
```

```python
def Solution(boxes):
    if len(boxes)==0: return 0
    if len(boxes)==1: return boxes[0][2]
    
    #sort boxes according to [L_i,W_i]
    boxes.sort(reverse=True)
    
    #HH[k]: highest height of stacked boxes ending at index k
    HH = [LWH[2] for LWH in boxes]
    for i in range(1,len(boxes)):
        options = [HH[j] for j in range(i) if boxes[j][0]>boxes[i][0] and boxes[j][1]>boxes[i][1]]
        HH[i] += max(options, default=0)

    return max(HH)
```

```python
Solution(boxes)
```
