# Brute Force Implementation

```python
def bf_locate_substring(s, t):
    """
    s: orignial string
    t: substring to locate
    """
    if len(s)==0: return -1
    if len(t)==0: return 0
    
    for i in range(len(s)):
        cur_head = s[i]
        if cur_head==t[0]:
            if s[i:i+len(t)]==t: return i
    return -1
```

# KMP Algorithm - Use Info from Pattern String "t"

```python
def get_next(pattern):
    next_arr = [0 for _ in range(len(pattern))]
    next_arr[0] = -1
    for i in range(2, len(pattern)):
        for j in range(i):
            if pattern[:j]==pattern[:i][-j:]:
                next_arr[i] = j
    return next_arr
```

```python
def kmp(string, pattern):
    next_arr = get_next(pattern)
    i, j = 0, 0
    while i<len(string) and j<len(pattern):
        if (j==-1) or (string[i]==pattern[j]):
            i += 1
            j += 1
        else:
            j = next_arr[j]
            
    if (j==len(pattern)):
        return i-j
    
    return -1
```

```python
kmp(s1, t1), kmp(s2, t2)
```
