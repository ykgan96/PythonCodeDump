# Upsampling

```python
# Creating Quarterly data
dates = pd.date_range(start='2016', periods=4, freq='Q')
data = range(1, 5)
quarterly = pd.Series(data=data, index=dates)
quarterly
```

```python
monthly = quarterly.asfreq('M') # to month-end frequency
monthly = monthly.to_frame('baseline') # to DataFrame
monthly
```

```python
# Differnet Upsampling fill methods
monthly['ffill'] = quarterly.asfreq('M', method='ffill') # bfill back fill
monthly['bfill'] = quarterly.asfreq('M', method='bfill') # ffill : forward fill
monthly['value'] = quarterly.asfreq('M', fill_value=0)
monthly
```

```python
quarterly.resample('M')
quarterly.asfreq('M').equals(quarterly.resample('M').asfreq())
```

# Join, Concat, Merge

1. merge() for combining data on common columns or indices
2. join() for combining data on a key column or an index
3. concat() for combining DataFrames across rows or columns

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df1
```

```python
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
df2
```

- module function
- You can use merge() anytime you want functionality similar to a database’s join operations
- When you want to combine data objects based on one or more keys, similar to what you’d do in a relational database, merge() is the tool you need. More specifically, merge() is most useful when you want to combine rows that share data.

```python
df_total = pd.merge(
    df1,
    df2,
    how='left', #outer/left/right
    on=['key'],
    #left_on=['key1','key2'],
    #right_on=['key1','key2'],
)
df_total
```

- a class method of pd.DataFrame
- join() is an instance method that lives on pd.DataFrame
- if key not specified, join by common column
- if key specified, join by key

```python
df1.join(df2,lsuffix='_caller', rsuffix='_other')
```

```python
df1.set_index('key').join(df2.set_index('key'))
```

```python
df1.join(df2.set_index('key'), on='key')
```

- with concatenation, your datasets are just stacked together along an axis


```python
pd.concat([df1,df2])
```
