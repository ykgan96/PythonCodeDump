```python
def pathSum(self, root, sum):
    """
    :type root: TreeNode
    :type sum: int
    :rtype: int
    """
    if root==None:
        return 0
    if root.left==None and root.right==None and root.val!=sum:
        return 0
        
    self.count = 0
    def path(node,presum,StartFromHere):
        if node==None:
            return
            
        if StartFromHere==True:
            if node.val==sum:
                self.count+=1
            path(node.left,node.val,False)
            path(node.right,node.val,False)
            path(node.left,0,True)
            path(node.right,0,True)
        else:
            if presum+node.val==sum:
                self.count+=1
            path(node.left,presum+node.val,False)
            path(node.right,presum+node.val,False)
                
    path(root,0,True)
    return self.count
```
