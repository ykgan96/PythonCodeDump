```python
def moveChars(initPos, speed):
    # Write your code here
    import math
    n = len(initPos)
    res = [initPos.replace('L','X').replace('R','X')]
    left, right = [], []
    for i in range(n):
        j = n-1-i
        if initPos[i]=='L':
            left.append(i)
        if initPos[j]=='R':
            right.append(j)

    if len(left)==0 and len(right)==0:
        return res
    elif len(left)!=0 and len(right)==0:
        t = math.ceil((max(left)+1)/speed)
    elif len(left)==0 and len(right)!=0:
        t =  math.ceil((n-1-min(right))/speed)
    else:
        t = max(math.ceil((max(left)+1)/speed), math.ceil((n-1-min(right))/speed))
    if t<=1:
        res.append('.'*n)
        return res

    for i in range(1,t+1):
        st = ['.' for i in range(n)]
        for m in range(len(left)):
            if left[m]!=-1000:
                if (left[m]-speed)<0:
                    left[m]=-1000
                else:
                    left[m]-=speed
        for k in range(len(right)):
            if right[k]!=-1000:
                if (right[k]+speed)>=n:
                    right[k]=-1000
                else:
                    right[k]+=speed
        for ind in left+right:
            if ind!=-1000:
                #print(ind)
                st[ind]='X'
        st = "".join(st)
        
        if st not in res:
            res.append(st)
            
    if res[-1]!='.'*n:
        res.append('.'*n)

    return res
```

```python
moveChars('..R....',2)
```

```python
moveChars('L............',3)
```
