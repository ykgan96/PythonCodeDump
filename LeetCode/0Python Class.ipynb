# No Init
Note: function must contain self

```python
class MyClass1:
    x = 2
    y = 5
    
    def f(self):
        return 'hello world'
```

```python
temp = MyClass1()
```

```python
temp.x, temp.y, temp.f()
```

# With Init

```python
class MyClass2:
    def __init__(self, name, sex):
        self.name = name
        self.sex = sex
    
    def introduce(self):
        return 'Name: ' + self.name + ', Sex: ' + self.sex
```

```python
t = MyClass2('GYK','Male')
```

```python
t.name, t.sex, t.introduce()
```

# Inheritance

```python
class ParentClass(object):
    def __init__(self, name, sex):
        self.name = name
        self.sex = sex
    
    def introduce(self):
        return 'Name: ' + self.name + ', Sex: ' + self.sex
```

```python
class SubClass(ParentClass):
    def __init__(self, name, sex, grade):
        ParentClass.__init__(self, name, sex)
        self.grade = grade
        
    def introduce(self):
        return 'Name: ' + self.name + ', Sex: ' + self.sex + ', Grade: ' + str(self.grade)
```

```python
t1 = SubClass('GYK','male',100)
```

```python
t1.name, t1.sex, t1.grade, t1.introduce()
```
