```python
def trans(s):
    ans = ''
    while len(s)>0:
        if len(s)==1:
            ans += '1'+s
            break
        cur,count = s[0],0
        while s[count]==s[count+1]:
            count+=1
            if count==len(s)-1 and s[count]==cur:
                break
        count+=1
        ans += str(count)+cur
        s = s[count:]
    return ans
```

```python
trans('312211')
```

```python
def CountAndSay(n):
    res=['1','11','21','1211','111221']
    if n<=5: return res[n-1]
        
    for i in range(5,n):
        last = res[i-1]
        cur = trans(last)
        res.append(cur)
    return res[-1]
```

```python
CountAndSay(7)
```
